```python
# Conecto a Mongodb 
client = pymongo.MongoClient('localhost',27017)
db = client['CCCba']

# Creo la colección del mismo nombre
collection=db['CCCba_collection']

# Cargo los datos del archivo csv
df=pd.read_csv('C:\\Users\\User\\Desktop\\TRABAJO FINAL MONGODB\\CCCba.csv')

# Genera un diccionario a partir de un dataframe Pandas
data=df.to_dict('records')

# Ahora inserto
collection.insert_many(data)
print("La información ha sido cargada exitosamente")
```

```python
sns.set(style = 'whitegrid')
sns.lmplot(x='Casos registrados Provincia de Córdoba (acumulado)',
           y='Fallecidos Provincia de Córdoba (acumulado)',
           data=df,
           fit_reg=True,
           scatter=True,
           ) 
 
```

```python
sns.lmplot(x='Fallecidos Provincia de Córdoba (acumulado)',
           y='Casos registrados Provincia de Córdoba (acumulado)',
           data=df,
           fit_reg=True,
           scatter=True,
           hue = 'Fallecidos Provincia de Córdoba (acumulado)'
          ) 
 
```
